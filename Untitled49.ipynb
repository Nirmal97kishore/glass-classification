```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import (train_test_split, KFold , StratifiedKFold, 
                                     cross_val_score, GridSearchCV, 
                                     learning_curve, validation_curve)
```

```python
df=pd.read_csv("glass.csv")
df.head()
features = df.columns[:-1].tolist()
```

```python
df.info()
```

```python
df.describe()
```

```python
for feat in features:
    skew = df[feat].skew()
    sns.distplot(df[feat], kde= False, label='Skew = %.3f' %(skew), bins=30)
    plt.legend(loc='best')
    plt.show()
```

```python
scaler=StandardScaler()
scaled_feature=scaler.fit_transform(df.drop(columns=["Type"]))
scaled=pd.DataFrame(scaled_feature,columns=df.columns[:-1])
```

```python
X_train,X_test,y_train,y_test=train_test_split(scaled,df['Type'],test_size=.3)
```

```python
Knn_euclidean=KNeighborsClassifier(n_neighbors=5,metric='euclidean')
Knn_euclidean.fit(X_train,y_train)
```

```python
y_pred_euclidean=Knn_euclidean.predict(X_test)
```

```python
Knn_manhattan=KNeighborsClassifier(n_neighbors=5,metric='manhattan')
Knn_manhattan.fit(X_train,y_train)
```

```python
y_pred_manhattan=Knn_manhattan.predict(X_test)
```

```python
acc = accuracy_score(y_test, y_pred_euclidean)
acc
```

```python
ac2=accuracy_score(y_test, y_pred_manhattan)
ac2
```
